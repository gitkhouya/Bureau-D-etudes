### 6. Sélection des caractéristiques

Nous allons créer une fonction pour tester le modèle RandomForestClassifier sur chaque caractéristique et afficher les métriques du modèle sur chaque liste de caractéristiques. 

```python
def feature_tester(feature_list):
    random_forest = RandomForestClassifier(n_estimators=200)
    Xtesting = accidents[feature_list]  # Features
    ytesting = accidents['new_grav']  # variable cible
    X_traintesting, X_testtesting, y_traintesting, y_testtesting = train_test_split(Xtesting, ytesting, test_size=0.2, random_state=42)
    random_forest.fit(X_traintesting,y_traintesting)
    Y_predtesting = random_forest.predict(X_testtesting)
    random_forest.score(X_testtesting, y_testtesting)
    acc_random_forest1 = round(random_forest.score(X_testtesting, y_testtesting) * 100, 2)
    sk_report = classification_report(
        digits=6,
        y_true=y_testtesting, 
        y_pred=Y_predtesting)
    print("Précision" , acc_random_forest1)
    print(sk_report)
```

#### 1. Application du filtre VarianceThreshold

```python
X
```

```python
from sklearn.feature_selection import VarianceThreshold

# seuil fixé à 87% pour la variance 
# c'est-à-dire que si 87% des données de la colonne sont identiques 
#(c'est-à-dire une faible variation), la colonne ne sera pas aussi utile
# dans la prédiction
thresh=(.85 * (1 - .85))
```

```python
# Fonction d'habillage pour identifier les caractéristiques à faible variance et les supprimer de la base de données 

def colonnes_faible_variance(dframe=None, columns=None,
                             skip_columns=None, thresh=0.0,
                             autoremove=False):
    try:
        # obtenir la liste de toutes les colonnes df originales
        all_columns = dframe.columns

        # supprimer `skip_columns`
        remaining_columns = all_columns.drop(skip_columns)

        # obtenir la longueur du nouvel index
        max_index = len(remaining_columns) - 1

        # obtenir les indices pour skip_columns
        skipped_idx = [all_columns.get_loc(column)
                       for column
                       in skip_columns]

        # ajuster l'emplacement de l'insertion en fonction du nombre de colonnes supprimées
        # (pour les emplacements d'insertion non nuls) afin de conserver les emplacements relatifs
        # les positions relatives intactes
        for idx, item in enumerate(skipped_idx):
            if item > max_index:
                diff = item - max_index
                skipped_idx[idx] -= diff
            if item == max_index:
                diff = item - len(skip_columns)
                skipped_idx[idx] -= diff
            if idx == 0:
                skipped_idx[idx] = item

        # obtenir les valeurs de  `skip_columns`
        skipped_values = dframe.iloc[:, skipped_idx].values

        # obtenir les valeurs de dataframe 
        X = dframe.loc[:, remaining_columns].values

        # instanciation de l'objet VarianceThreshold
        vt = VarianceThreshold(threshold=thresh)

        # adapter vt aux données
        vt.fit(X)

        # obtenir les indices des caractéristiques qui sont conservées
        feature_indices = vt.get_support(indices=True)

        # supprimer les colonnes à faible variance de l'index
        feature_names = [remaining_columns[idx]
                         for idx, _
                         in enumerate(remaining_columns)
                         if idx
                         in feature_indices]

        # obtenir les colonnes à supprimer
        removed_features = list(np.setdiff1d(remaining_columns,
                                             feature_names))
        print("A trouvé {0} colonnes à faible variance."
              .format(len(removed_features)))

        # supprimer les colonnes
        if autoremove:
            print("Suppression des caractéristiques à faible variance.")
            # remove the low-variance columns
            X_removed = vt.transform(X)

            print("Réassemblage de la base de données (avec des données à faible variance) "
                  "fonctionnalités supprimées).")
            # re-assemble the dataframe
            dframe = pd.DataFrame(data=X_removed,
                                  columns=feature_names)

            # add back the `skip_columns`
            for idx, index in enumerate(skipped_idx):
                dframe.insert(loc=index,
                              column=skip_columns[idx],
                              value=skipped_values[:, idx])
            print("Suppression réussie des colonnes à faible variance.")

        # do not remove columns
        else:
            print("Aucune modification n'a été apportée à la base de données.")

    except Exception as e:
        print(e)
        print("Impossible de supprimer les caractéristiques à faible variance.")
        pass

    return dframe, removed_features
```

### Traitement les données déséquilibrées

#### Métrique populaire
La métrique la plus simple pour mesurer la classification est la précision de base, c'est-à-dire le rapport entre les prédictions correctes et le nombre total d'échantillons dans l'ensemble de données. Toutefois, dans le cas de classes déséquilibrées, cette mesure peut être trompeuse, car les scores élevés n'indiquent pas la capacité de prédiction pour la classe minoritaire. On peut avoir une précision de 99 % mais une faible capacité de prédiction pour la classe qui nous intéresse vraiment (par exemple, la détection d'anomalies où les anomalies sont des classes rares dans un ensemble de données). 

C'est pourquoi nous commençons par calculer d'autres mesures de performance telles que la AUC, recall et F1 scores

**AUC - ROC** est une mesure de performance pour les problèmes de classification à différents seuils. La courbe ROC est une courbe de probabilité et la courbe AUC représente le degré ou la mesure de séparabilité. Elle indique dans quelle mesure le modèle est capable de faire la distinction entre les classes. Plus la SSC est élevée, plus le modèle est capable de prédire les 0 comme des 0 et les 1 comme des 1.

**recall** correspond au pourcentage de résultats pertinents correctement classés par l'algorithme.

**F1 score** peut être interprété comme une moyenne pondérée de la précision et du rappel, où un score F1 atteint sa meilleure valeur à 1 et sa pire valeur à 0.

## Méthodes de suréchantillonnage : SMOTE


SMOTE utilise un algorithme de nearest neighbors pour générer de nouvelles données synthétiques que nous pouvons utiliser pour former notre modèle.
Une fois encore, il est important de générer de nouveaux échantillons uniquement dans l'ensemble de training afin de s'assurer que notre modèle se généralise bien à des données inédites.

## Méthodes de Sous-échantillonnage : Tomek Link

Le suréchantillonnage et le sous-échantillonnage sélectionnent les exemples de manière aléatoire afin d'ajuster leurs proportions. D'autres approches examinent attentivement l'espace des instances et décident de ce qu'il convient de faire en fonction de leur voisinage.

Pour plus de détails sur le Kappa de Cohen, nous vous renvoyons à la référence suivante [Tomek Link](https://www.researchgate.net/publication/360157856_Tomek_Link_and_SMOTE_Approaches_for_Machine_Fault_Classification_with_an_Imbalanced_Dataset)

```python
from imblearn.under_sampling import TomekLinks
```

```python
majority_class = 0
tl = TomekLinks(sampling_strategy='majority', n_jobs=100)
X_tl, y_tl = tl.fit_resample(X_train, y_train)
```

Avant le sous-échantillonnage

```python
# Nombre de modalité dans l'ensemble d'apprentissage avant le sous-échantillonnage

print("Avant le suréchantillonnage, nombre d'occurence de 1: {}".format(sum(y_train==1)))
print("Avant le suréchantillonnage, nombre d'occurence de 0 : {} \n".format(sum(y_train==0)))
```

 Après  le sous-échantillonnage

```python
# Nombre de modalité dans l'ensemble d'apprentissage après le sous-échantillonnage

print("Après le suréchantillonnage, nombre d'occurence de 1: {}".format(sum(y_tl==1)))
print("Après le suréchantillonnage, nombre d'occurence de 0 : {} \n".format(sum(y_tl==0)))
```

Par conséquent, nous pouvons nous attendre aux mêmes résultats que ceux observés ci-dessus (en utilisant l'ensemble de données original déséquilibré), lorsque nous utilisons les différents algorithmes pour former et tester le modèle. En tout cas, le sous-échantillonnage n'est peut-être pas la meilleure solution dans notre scénario. En outre, en sous-échantillonnant, nous courons également le risque de perdre des informations potentiellement significatives.

À des fins de test pour prouver l'affirmation ci-dessus, nous appliquerons les données sous-échantillonnées à la logistique et nous comparerons le rappel et le score f1 à ceux obtenus avant le sous-échantillonnage.
